```python
import pandas as pd
import numpy as np
```

```python
url= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
```

```python
df=pd.read_csv(url)
```

```python
df.info()
```

```python
df.head()
```

```python
print(df.head)
```

```python
headers=[ "Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"]
```

```python
df.columns= headers
df.head(5)
```

```python
df.tail(25)
```

```python
df.replace('?', np.nan, inplace=True)
print(df.dtypes)
```

```python
print(df.describe())
```

```python
print(df.describe(include='all'))
```

```python
df['Screen_Size_inch'] = pd.to_numeric(df['Screen_Size_inch'], errors='coerce')
print(df.dtypes)
```

```python
df['Weight_kg'] = pd.to_numeric(df['Weight_kg'], errors='coerce')
print(df.dtypes)
```

```python
print('Missing values after type conversion:')
print(df.isnull().sum())
```

```python
print('\nUpdated descriptive statistics:')
print(df.describe(include='all'))
```
